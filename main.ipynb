Basic commands for the bot

```python
import telebot

bot=telebot.TeleBot(Api_token)

@bot.message_handler(commands=['start'])
def send_welcome(message):
    bot.reply_to(message, "Welcome to the ERM AXUALLAY bot!\nThis bot helps you fact check any unverified claims that you may hear.\nType /help to see the available commands.")

@bot.message_handler(commands=['help'])
def send_help(message):
    bot.reply_to(message, "Here are the available commands:\n/start - Start the bot\n/help - Show this help message")

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    bot.reply_to(message, message.text)

bot.polling()
```
