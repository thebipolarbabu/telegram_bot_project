Retrieve articles using duckduckgo search text method - no filter applied yet

```python
import pandas as pd
from ddgs import DDGS
import newspaper
from newspaper import Article
```

```python
results = DDGS().text(query="galgotias impact ai summit", region="in-en",safesearch= "off", timelimit="7d", max_results=30)
df=pd.DataFrame(results)
df
```

```python
for url in df['href']:
    try:
        article = Article(url)
        article.download()
        article.parse()
        df.loc[df['href'] == url, 'text'] = article.text
    except Exception as e:
        print(f"Error processing URL {url}: {e}")
df
```
