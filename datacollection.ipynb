Retrieve articles using duckduckgo search text method - no filter applied yet

```python
import pandas as pd
from ddgs import DDGS
import newspaper
from newspaper import Article
```

```python
results = DDGS().text(query="galgotias impact ai summit", region="in-en",safesearch= "off", timelimit="7d", max_results=30)
df=pd.DataFrame(results)
df
```

```python
for url in df['href']:
    try:
        article = Article(url)
        article.download()
        article.parse()
        df.loc[df['href'] == url, 'text'] = article.text
    except Exception as e:
        print(f"Error processing URL {url}: {e}")
df
```

Data preparation

```python
df.dropna(inplace=True)
df
```

pre-processing data into tokenized sentences

```python
all_text = ' '.join(df['text'])
all_text = all_text.lower()
all_text
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(all_text)
sentences = pd.DataFrame([sent.text for sent in doc.sents])
sentences
```

Comparision of tokenized sentences with a claim to check validity (sample output)

```python
claim = nlp("the galgotias university made an ai powered robot dog in their centre of excellence")
matches = sentences[0].apply(lambda x: nlp(x).similarity(claim))
matches
```

```python
#tried to find the most similar sentences to the claim but they do not seem relevant when viewd manually. This could be because the claim is very specific and the sentences in the collected data may not have enough context to match it accurately. The similarity score is based on the overall meaning of the sentences, so it may not capture specific details that are crucial for verifying the claim.
matches.sort_values(ascending=False)
indices=matches.sort_values(ascending=False).head(15).index
print(indices)
print(sentences.iloc[indices])
```

```python
if(matches.mean()>0.75):
    print("The claim is likely to be true based on the collected data.")
else:
    print("The claim is likely to be false based on the collected data.")
```
